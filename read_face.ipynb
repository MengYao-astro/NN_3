```python
from PIL import Image
import numpy as np
import os
import matplotlib.pyplot as plt

baldpath = './face_images/bald/'
hairpath = './face_images/hair/'
baldlist = os.listdir(baldpath)
hairlist = os.listdir(hairpath)
face=[]
label=[]
for i in baldlist:
    try:
        image =  Image.open(baldpath+i).resize((128,128))
        data = np.array(image)/255.
        face.append(data)
        label.append(0)
    except:
        print(i+' is not a image file, skipping it')
for j in hairlist:
    try:
        image =  Image.open(hairpath+j).resize((128,128))
        data = np.array(image)/255.
        face.append(data)
        label.append(1)
    except:
        print(j+' is not a imgae file, skipping it')
face=np.array(face)
label=np.array(label)
#np.save('face',face)
#np.save('label',label)
```

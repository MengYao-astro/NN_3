```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL


number = 100 #predictions.shape[0]
dir1 = 'features/'
fig = plt.figure(figsize=(10,10))
for i in range(number):
    name_im = dir1 + 'feature' + str(i) + '.npy'
    image = np.load(name_im)
    plt.subplot(10, 10, i+1)
    plt.imshow(image, cmap='gray')
    plt.axis('off')
    name_fig = dir1+'image.png'
    #plt.savefig(name_fig)
    #plt.close()
plt.show()
```

```python
noise = np.load(dir1+'noise_vector.npy')
noise_bald_mean = (noise[95,:]+noise[77,:]+noise[14,:])/3.
noise_glasses_mean = (noise[7,:]+noise[30,:]+noise[42,:])/3.
noise_hair_mean = (noise[46,:]+noise[51,:]+noise[56,:])/3.

noise_hair_feature = noise_hair_mean - noise_bald_mean
noise_glasses_feature = noise_glasses_mean - noise_hair_mean

noise_blad14_hair = (noise[14,:]+noise_hair_feature)
noise_blad14_glasses = (noise[14,:] + noise_glasses_feature)
noise_hair51_glasses = (noise[51,:] + noise_glasses_feature)

noise_edit = np.zeros((6,100))
noise_edit[0,:] = noise_blad14_hair
noise_edit[1,:] = noise_blad14_glasses
noise_edit[2,:] = noise_hair51_glasses
noise_edit[3,:] = noise_bald_mean
noise_edit[4,:] = noise_glasses_mean
noise_edit[5,:] = noise_hair_mean
noise_edit.shape
np.save('noise_edit',noise_edit)
```
